```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
DATASET_PATH = '../creditcard.csv'
```

```python
df = pd.read_csv(DATASET_PATH, sep=',')
df.shape
```

```python
df.head(4)
```

```python
df.describe()
```

```python
result = df['Class'].value_counts() # сохраним получившийcя объект Series в переменную result
```

```python
plt.style.use('fivethirtyeight') 
```

```python
result.plot(kind="bar")
plt.show()
```

```python
result.plot(kind="bar", logy=True)
plt.show()
```

```python
class0 = df.loc[df['Class'] == 0, 'V1']
```

```python
class1 = df.loc[df['Class'] == 1, 'V1']
```

```python
plt.hist(class0, bins=20, alpha = 0.5, color = 'grey', density = True)
plt.hist(class1, bins=20, alpha = 0.5, color = 'red', density = True)
plt.xlabel("V1")
plt.axis([-30, 2.5, 0, 0.23])
plt.legend(labels=["Class 0", "Class 1"])


plt.show()
```
